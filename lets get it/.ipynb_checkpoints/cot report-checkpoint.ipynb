```python
def get_rates(pair1, timeframe, x):
    pair1 = pd.DataFrame(mt5.copy_rates_from_pos(pair1, timeframe, 0, x))
    pair1['time'] = pd.to_datetime(pair1['time'], unit = 's')
    return pair1[['time','open', 'high', 'low', 'close']].set_index('time')
def get_rates_w(pair1, timeframe, x):
    
    pair1 = pd.DataFrame(mt5.copy_rates_from_pos(pair1, timeframe, 0, x))
    
    pair1['time'] = pd.to_datetime(pair1['time'], unit='s')
    pair1 = pair1[['time', 'open', 'high', 'low', 'close']].set_index('time')
    
    # Resample to weekly data starting from the day after COT report release
    weekly_data = pair1.resample('W-TUE', label='left').agg({
        'open': 'first',
        'high': 'max',
        'low': 'min',
        'close': 'last'
    })

    return weekly_data
```

```python
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

# Initial VIF calculation
vif_data = calculate_vif(X)

# Iteratively remove features with the highest VIF
while vif_data['VIF'].max() > 4:  # Assuming 10 is your threshold
    highest_vif_feature = vif_data.sort_values('VIF', ascending=False)['feature'].iloc[0]
    X = X.drop(columns=[highest_vif_feature])
    vif_data = calculate_vif(X)

print(vif_data)
```
